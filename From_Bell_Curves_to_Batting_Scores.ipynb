⚡ ***Normal Distribuition***

**Normal Distribution, also known as the Gaussian Distribution or Bell Curve, is a statistical distribution in which most values are concentrated around the mean (average), while fewer values occur farther away from the mean.**

It forms a symmetric bell-shaped curve where:

• Mean = Median = Mode

• Data near the center occurs more frequently

• Extreme values are rare

• The total area under the curve equals 1




***In this project, we use Normal Distribution to understand how cricket batting scores are distributed and whether player performances follow a predictable statistical pattern.***

📌 ***Parameters of Normal Distribution***

• μ (Mu)  → Mean of the distribution

• σ (Sigma) → Standard deviation

• σ² → Variance


⚡ ***Skewness***

**Skewness is a statistical measure that describes the asymmetry of a data distribution around its mean.**

It helps us understand whether the data is:
- Symmetrically distributed
- Skewed toward higher values
- Skewed toward lower values

**In a perfectly normal distribution, the distribution is completely symmetric and skewness is equal to 0.**

📌***Types of Skewness:***

• Positive Skewness:
  The right tail is longer.
  Few unusually high values pull the distribution toward the right.

• Negative Skewness:
  The left tail is longer.
  Few unusually low values pull the distribution toward the left.

• Zero Skewness:
  The distribution is approximately symmetric.

```python
plt.figure(figsize=(10,5))

sns.histplot(
    data["Runs"],
    kde=True,
    bins=15
)

plt.title("Normal Distribution of Cricket Scores")
plt.xlabel("Runs")
plt.ylabel("Frequency")

plt.savefig("normal_distribution.png")
plt.show()
```

• **The distribution of runs appears approximately bell-shaped, showing characteristics of a normal distribution.**


• **Very high and very low run values occurred less frequently compared to average runs.**


• **The KDE curve provided a smoother understanding of how batting runs are distributed.**

• **The visualization helps understand batting consistency and variability in player performances.**

⚡***Standard Normal Variate (Z-Score)***

**Standard Normal Variate, also known as Z-Score, is a statistical measure used to determine how far a value is from the mean in terms of standard deviations.**


It helps standardize data so different observations can be compared on the same scale.

A Z-score tells us:
- Whether a value is above or below the mean
- How unusual or common an observation is

🧠***Interpretation:***

• Z = 0  → Value is exactly at the mean

• Z > 0 → Value is above the mean

• Z < 0 → Value is below the mean

📌 ***Parameters***

• X  → Individual value

• μ  → Mean of the data

• σ  → Standard deviation

🧠 ***Why is Z-Score Important?***

• **Helps identify unusual performances**

• **Detects outliers in the dataset**

• **Standardizes values for comparison**

• **Converts normal distribution into standard normal distribution**

• **Used in probability and statistical analysis**

📌 ***Calculate Z-Score***

```python
data["Z_Score"] = (
    (data["Runs"] - data["Runs"].mean())
    / data["Runs"].std()
)

data.head()
```

📌***Z-Score Visualization***

```python
plt.figure(figsize=(10,5))

sns.histplot(
    data["Z_Score"],
    kde=True,
    bins=15
)

plt.title("Distribution of Z-Scores for Cricket Runs")
plt.xlabel("Z-Score")
plt.ylabel("Frequency")

plt.axvline(0, linestyle="--")

plt.savefig("z_score.png")
plt.show()
```

##📊 Insights

• **Most Z-score values are concentrated around 0, indicating that most cricket runs are close to the average.**

• **Positive Z-scores represent innings above the average run value.**

• **Negative Z-scores represent innings below the average run value.**

• **Extremely high or low Z-score values indicate unusual batting performances.**

• **The distribution helps identify consistency and rare performances among players.**

⚡***Empirical Rule (68–95–99.7 Rule)***

**The Empirical Rule states that in a normal distribution:**

• **Approximately 68% of the data lies within 1 standard deviation from the mean.**

• **Approximately 95% of the data lies within 2 standard deviations from the mean.**

• **Approximately 99.7% of the data lies within 3 standard deviations from the mean.**

***This rule helps us understand how data is distributed around the average value.***

📌***Applying Empirical Rule on Cricket Runs***

```python
# Mean and Standard Deviation
mean = data["Runs"].mean()
std = data["Runs"].std()

# 1 Standard Deviation
within_1_std = data[
    (data["Runs"] >= mean - std) &
    (data["Runs"] <= mean + std)
]

# 2 Standard Deviations
within_2_std = data[
    (data["Runs"] >= mean - 2*std) &
    (data["Runs"] <= mean + 2*std)
]

# 3 Standard Deviations
within_3_std = data[
    (data["Runs"] >= mean - 3*std) &
    (data["Runs"] <= mean + 3*std)
]

print("Within 1 STD:", len(within_1_std))
print("Within 2 STD:", len(within_2_std))
print("Within 3 STD:", len(within_3_std))
```

##📊 Insights

• **Most cricket run values are concentrated near the average performance range.**

• **Very few innings fall far away from the mean, showing the rarity of extremely high or low performances.**

• **The empirical rule helps measure batting consistency and variability.**

• **Observing how much data falls within standard deviation ranges helps evaluate whether the distribution behaves approximately normally.**

⚡***Symmetricity in Normal Distribution***

**Symmetricity means that a distribution is evenly balanced around its center (mean).**

In a perfectly symmetric distribution:

• Left side of the curve mirrors the right side

• Mean = Median = Mode

• Both tails are equal in shape and length


In real-world cricket data:

• Perfect symmetry is rare

• Small deviations occur due to exceptional high or low performances

⚡***Symmetry Check***

```python
mean_runs = data["Runs"].mean()
median_runs = data["Runs"].median()

print("Mean:", mean_runs)
print("Median:", median_runs)
```

📌***Visualization***

```python
plt.figure(figsize=(10,5))

sns.histplot(data["Runs"], kde=True, bins=15)

plt.axvline(data["Runs"].mean(), color="red", linestyle="--", label="Mean")
plt.axvline(data["Runs"].median(), color="green", linestyle="--", label="Median")

plt.title("Symmetry in Cricket Runs Distribution")
plt.xlabel("Runs")
plt.ylabel("Frequency")

plt.legend()

plt.savefig("symmetry.png")
plt.show()
```

• **The distribution of cricket runs is approximately symmetric but not perfectly balanced.**

• **Mean and median are close, indicating near-symmetric behavior.**

• **Slight deviations occur due to a few high-scoring innings.**

• **Real-world cricket data rarely shows perfect symmetry due to performance variability.**

• **Symmetry helps us understand whether average performance truly represents the dataset.**
